```python
def print_board(board):
    for row in board:
        print(" | ".join(row))
        print("-" * 9)

def check_winner(board, player):
    
    for i in range(3):
        if all([cell == player for cell in board[i]]) or all([board[j][i] == player for j in range(3)]):
            return True
    
    if all([board[i][i] == player for i in range(3)]) or all([board[i][2 - i] == player for i in range(3)]):
        return True
    return False

def is_full(board):
    return all(cell != " " for row in board for cell in row)

def get_move():
    while True:
        try:
            row, col = map(int, input("Enter row and column (0-2, space-separated): ").split())
            if 0 <= row < 3 and 0 <= col < 3:
                return row, col
            else:
                print("Invalid input. Try again.")
        except ValueError:
            print("Invalid input. Enter two numbers.")

def tic_tac_toe():
    board = [[" " for _ in range(3)] for _ in range(3)]
    players = ["X", "O"]
    turn = 0

    while True:
        print_board(board)
        print(f"Player {players[turn]}'s turn")
        
        row, col = get_move()
        if board[row][col] == " ":
            board[row][col] = players[turn]
            if check_winner(board, players[turn]):
                print_board(board)
                print(f"Player {players[turn]} wins!")
                break
            elif is_full(board):
                print_board(board)
                print("It's a tie!")
                break
            turn = 1 - turn
        else:
            print("Spot already taken, try again.")

tic_tac_toe()
```
